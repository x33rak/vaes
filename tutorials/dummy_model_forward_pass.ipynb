```python
import torch
from torch import nn
from torch.functional import F
```

```python
batch_size = 256
d = 20
x = torch.randn(size=(batch_size, 1, 28, 28))
x_maybe = torch.randn(size=(batch_size, 28, 28))
```

```python
class VAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(784, d**2),
            nn.ReLU(),
            nn.Linear(d**2, d*2)
        )
        self.decoder = nn.Sequential(
            nn.Linear(d, d**2),
            nn.ReLU(),
            nn.Linear(d**2, 784),
            nn.Sigmoid()
        )
    def reparameterize(self, mu, logvar):
        std = logvar.mul(0.5).exp_()
        eps = std.new_empty(std.size()).normal_()
        return eps.mul_(std).add_(mu)
        
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        mu_logvar = self.encoder(x.view(-1, 784)).view(-1, 2, d)
        print(mu_logvar.shape)
        
        mu = mu_logvar[:, 0, :]
        print(mu.shape)
        
        logvar = mu_logvar[:, 1, :]
        print(logvar.shape)
        
        z = self.reparameterize(mu, logvar)
        print(z.shape)

        final = self.decoder(z)
        print(final.shape)
        
        return final, mu, logvar
```

```python
model = VAE()
out_val, _, _ = model(x.view(-1, 784))
```
